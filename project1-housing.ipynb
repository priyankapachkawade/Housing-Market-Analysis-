```python
#.text make a function to add labels for x,y ( x and y are the argument 
# for every x y is label for x using plt.text
import matplotlib.pyplot as plt

def add_labels(x_values, y_values, labels):
    """
    Add labels next to each point on a plot.

    Parameters:
    - x_values: List or array of x coordinates.
    - y_values: List or array of y coordinates.
    - labels: List or array of labels corresponding to each (x, y) pair.

    Returns:
    - None (plots the points with labels).
    """
    # Check if the lengths of x_values, y_values, and labels are the same
    if len(x_values) != len(y_values) or len(x_values) != len(labels):
        raise ValueError("Lengths of x_values, y_values, and labels must be the same.")

    # Plot the points
    plt.scatter(x_values, y_values)

    # Add labels for each point
    for x, y, label in zip(x_values, y_values, labels):
        plt.text(x, y, label, fontsize=12, ha='left', va='bottom')

    # Show plot
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Plot with Labels')
    plt.grid(True)
    plt.show()

# Example usage:
x_data = [1, 2, 3, 4, 5]
y_data = [10, 20, 15, 25, 30]
labels = ['Point 1', 'Point 2', 'Point 3', 'Point 4', 'Point 5']

add_labels(x_data, y_data, labels)
```
